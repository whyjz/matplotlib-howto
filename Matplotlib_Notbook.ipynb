```python
import numpy as np
import matplotlib.gridspec as gridspec
from datetime import datetime
gs1 = gridspec.GridSpec(1, 4)
gs1.update(left=0.05, right=0.95, wspace=0)
ax1 = plt.subplot(gs1[0, 0])
ax2 = plt.subplot(gs1[0, 1:], sharey=ax1)
ax1.text(datetime(2002, 1, 1), 22, 'Terminus speed  ', bbox={'facecolor':np.array([208, 46, 98]) / 256.,  'alpha':0.75, 'pad':10})
ax1.text(datetime(2002, 1, 1), 19, 'Coastline -8  km', bbox={'facecolor':np.array([238, 97, 15]) / 256.,  'alpha':0.75, 'pad':10})
ax1.text(datetime(2002, 1, 1), 16, 'Coastline -4  km', bbox={'facecolor':np.array([203, 234, 25]) / 256., 'alpha':0.75, 'pad':10})
ax1.text(datetime(2002, 1, 1), 13, 'Coastline (1988)', bbox={'facecolor':np.array([41, 179, 191]) / 256., 'alpha':0.75, 'pad':10})
ax1.text(datetime(2002, 1, 1), 10, 'Coastline +4  km', bbox={'facecolor':np.array([25, 88, 234]) / 256.,  'alpha':0.75, 'pad':10})
ax1.text(datetime(2002, 1, 1), 7, 'Coastline +8  km', bbox={'facecolor':np.array([87, 46, 184]) / 256.,  'alpha':0.75, 'pad':10})
ax1.text(datetime(2002, 1, 1), 4,  'Coastline +12 km', bbox={'facecolor':np.array([210, 34, 208]) / 256., 'alpha':0.75, 'pad':10})

ax2.plot(datetime(2013, 2, 1), 21, marker='^', markersize=8, linewidth=2, markerfacecolor='w')
ax2.text(datetime(2013, 3, 1), 21, 'WV')
ax2.plot(datetime(2013, 2, 1), 19, marker='o', markersize=6, linewidth=2, markerfacecolor='w')
ax2.text(datetime(2013, 3, 1), 19, 'LS7/8 with same Path/Row')
ax2.plot(datetime(2013, 2, 1), 17, marker='D', markersize=4, linewidth=1, markerfacecolor='w')
ax2.text(datetime(2013, 3, 1), 17, 'LS7/8 with different Paths/Rows')

plt.setp(ax2.get_yticklabels(), visible=False)
#plt.setp(ax1, fontsize=10)

#ax1.set_xticklabels(size=14)
ax1.set_xlim(datetime(1999, 1, 1), datetime(2013, 1, 1))
ax1.set_yticks((0., 5., 10., 15., 20., 25., 30.))
ax2.set_xlim(datetime(2013, 1, 1), datetime(2016, 6, 1))
```
